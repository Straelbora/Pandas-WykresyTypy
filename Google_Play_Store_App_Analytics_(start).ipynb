# Extracting Nested Data from a Column

**Challenge**: How many different types of genres are there? Can an app belong to more than one genre? Check what happens when you use .value_counts() on a column with nested values? See if you can work around this problem by using the .split() function and the DataFrame's [.stack() method](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.stack.html).


```python
df_apps_clean.value_counts('Genres')
```

Hey, don't feel bad. It took me a while to get the concept too.. I'd say "it's poorly explained" but it's worst than that, since it's not explained at all..



I'll do my best to explain it step by step, hopefully it helps anyone!



First, I'll start by getting a DataFrame with only the apps with a Genre that contain "Education", simply because it's easier to visualise what's going on (for me at least)

df_apps_multiple_genres = df_apps_clean[df_apps_clean["Genres"].str.contains("Education")]



Then lets count the how many apps we have in each Genre using the .value_counts() function

df_apps_multiple_genres.Genres.value_counts()

Well, that doesn't look ok, does it? To take an example, there's 1 app with the Genre "Tools;Education".. But "Tools;Education" it's not a Genre, they are 2! So that app should be counted as 1 for "Tools" AND 1 for "Education", since a single app can have 2 genres.



So how we fix this? It takes 2 steps.



Step 1 - SPLIT:  We need to separate the Genres. That's where split() comes in. We can see that the data uses ";" to separate the Genres, so we will pass that as the argument for split.

separated_genres_df = df_apps_multiple_genres.Genres.str.split(";")

That's not great either, is it? The text was separated (now each Genre is an array of 1 or 2 values), but they are still in the same column, so we are in the same situation. However, split() has an optional argument expand that when it's set to True, instead of creating an array in the same column, will create new columns (and returns a dataframe.. It works a bit as the "Text to Columns" functionality in Excel, if you're familiar with it):

separated_genres_df = df_apps_multiple_genres.Genres.str.split(";", expand=True)



So now we're getting somewhere.. We have 2 columns (that are called "0" and "1" but you can think of them as "Main Genre" and "Additional Genre".. Notice how some values have None there (they already had 1 segment) but the ones that had 2 segments, now have 1 in each column.

Cool, but how do we count them?... One option would be call .value_counts() in the column "0", then do the same for column "1", and add up the values somehow... But there's a better option



Step 2 - STACK: Wouldn't it be easier to count if we could put column "1" below column "0"? In other words, if we could stack them? Surprise, surprise, that's exactly what the stack() function does!

stacked_genres = separated_genres_df.stack()

Notice how stack() gave us back a Series from a DataFrame, since it took 2 columns and made it one. Also notice how for the cases that only had one Genre (in red), nothing changed really, they just have 1 row.. However, for those with 2 Genres (like index 10244, in green) now they have 2 "sub-index" (I guess?) that correspond to the columns "0" and "1" from the separated_genres_df  .. Important, where you see the values "0" and "1" is NOT a column! The 0/1 values are also indexes, as I mentioned before, we're dealing with a Series here. And because it's a Series, we can FINALLY call the .value_counts() on it and get the count by individual Genre

stacked_genres.value_counts()



If you compare this with my first screenshot, notice how every row now has only ONE Genre, for instance "Tools" doesn't have the ";Education" part anymore, which ended up on the "Education" Genre, as it should be (same applies for all the others, that's why the number of "Education" is much higher)



We did it! Yay🎉🎉

```python
# Split the strings on the semi-colon and then .stack them.
# Dzielimy podwójne gatunki rozdzielone ; w jednej kolumnie i otrzymujemy listę,
# np. [Educational, Pretend Play] lub [Education]
# Jeśli dodamy parametr expand=True, te wartości z listy rozbijane są na oddzielne kolumny,
# np.     0       |     1
#     ---------------------------
#     Education   |    None
#     Educational |   Pretend Play
#
# .stack() pozwala nam połączyć znów te kolumny w jedną, ale z dodatkowymi indeksami:
#    0  Education
#   ----------------
#    0 Educational
#    1 Pretend Play
stack = df_apps_clean.Genres.str.split(';', expand=True).stack()
print(f'We now have a single column with shape: {stack.shape}')
num_genres = stack.value_counts()
print(f'Number of genres: {len(num_genres)}')
```

# Grouped Bar Charts: Free vs. Paid Apps per Category

```python
# df_apps_clean.Type.value_counts()
# as_index=False wypycha Category jako wartość w kolumnie a nie jako index
# bez tego mielibyśmy wynik:
#    Category    | Type | App
# ----------------------------
# ART_AND_DESIGN  Free  |  58
#                 Paid  |   3
# -----------------------------
df_free_vs_paid = df_apps_clean.groupby(['Category', 'Type'], as_index=False).agg({'App': pd.Series.count})
df_free_vs_paid.head()
```
